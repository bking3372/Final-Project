```python
import pandas as pd
import numpy as np
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
database_path = "Resources/FPA_FOD_20170508.sqlite"
```

```python
# Create Engine
engine = create_engine(f"sqlite:///{database_path}")
conn = engine.connect()
```

```python
# Query All Records in the the Database
data = pd.read_sql("SELECT * FROM Fires", conn)
```

```python
data.head()
```

```python
col_list = ['FOD_ID', 'SOURCE_SYSTEM_TYPE', 'FIRE_NAME', 'COMPLEX_NAME', 'FIRE_YEAR', 'DISCOVERY_DATE', 'DISCOVERY_DOY', 'DISCOVERY_TIME', 'STAT_CAUSE_CODE', 'STAT_CAUSE_DESCR', 'CONT_DATE', 'CONT_TIME', 'FIRE_SIZE', 'FIRE_SIZE_CLASS', 'LATITUDE', 'LONGITUDE', 'OWNER_CODE', 'OWNER_DESCR', 'STATE', 'COUNTY', 'FIPS_CODE', 'FIPS_NAME']
df = data[col_list]
```

```python
df.head(10)
```

```python
df.dtypes
```

```python
df = df.dropna(subset=['CONT_DATE'])
```

```python
len(df)
```

```python
df.head()
```

```python
df.loc[:,'DURATION'] = df['CONT_DATE'] - df['DISCOVERY_DATE']
```

```python
df["DURATION"].describe().astype(np.int64)
```

```python
import datetime as dt
```

```python
df['DATE'] = pd.to_datetime(df['DISCOVERY_DATE'] - pd.Timestamp(0).to_julian_date(), unit='D')
df.head()
```

```python
df['MONTH'] = pd.DatetimeIndex(df['DATE']).month
df['DAY_OF_WEEK'] = df['DATE'].dt.weekday
df.head()
```

```python
#export final file
output_data_file = "Resources/wildfires.csv"
```

```python
df.to_csv(output_data_file, index=False, header=True)
```
